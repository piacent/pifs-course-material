```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
N = 1000000

x1 = st.uniform.rvs(loc = 0, scale = 1, size = N)
x2 = st.uniform.rvs(loc = 0, scale = 1, size = N)

r1 = np.sqrt(-2. * np.log(1.-x1)) * np.cos(2. * np.pi * x2)
r2 = np.sqrt(-2. * np.log(1.-x1)) * np.sin(2. * np.pi * x2)

xline = np.linspace(-3, 3, 1000)
yline = st.norm.pdf(xline)

plt.hist(r1, bins = 30, density = True, color = 'C2', label = 'p(r1)')
plt.plot(xline, yline, color = 'black', linestyle = '--', label = 'N(mu = 0, sigma = 1)')
plt.ylabel('f(r1)')
plt.xlabel('r1')
plt.legend()
plt.show()



plt.hist(r2, bins = 30, density = True, color = 'C2', label = 'p(r2)')
plt.plot(xline, yline, color = 'black', linestyle = '--', label = 'N(mu = 0, sigma = 1)')
plt.ylabel('f(r2)')
plt.xlabel('r2')
plt.legend()
plt.show()
```
