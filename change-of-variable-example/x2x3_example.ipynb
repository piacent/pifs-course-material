```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
N = 100000

x = st.uniform.rvs(loc = -1, scale = 2, size = N)

y = x**2

y2 = x**3

xline = np.linspace(0.001, 1, 100000)
yline = 1. / 2. / np.sqrt(xline)

xline2 = np.linspace(-1.0, 1.0, 100000)
yline2 = 1. / 6. / (xline2**2)**(1/3)

plt.hist(x, bins = 30, density = True)
plt.hlines(1./2., -1, 1, color = 'red', linestyle = '--')
plt.ylabel(r'$f_X$(x)')
plt.xlabel('x')
plt.show()

plt.hist(y, bins = 30, density = True, color = 'C1')
plt.plot(xline, yline, color = 'red', linestyle = '--')
plt.hlines(1./2., 0, 1, color = 'black', linestyle = '--')
plt.ylabel(r'$f_Y$(y)')
plt.title(r'y = $x^2$')
plt.xlabel('y')
plt.ylim(0, 3)
plt.show()

plt.hist(y2, bins = 30, density = True, color = 'C2')
plt.plot(xline2, yline2, color = 'red', linestyle = '--')
plt.hlines(1./6., -1, 1, color = 'black', linestyle = '--')
plt.ylabel(r'$f_Y$(y)')
plt.title(r'y = $x^3$')
plt.xlabel('y')
plt.ylim(0, 3)
plt.show()
```
